```python
import pandas as pd
filepath = "imdb-movies-dataset.csv"
df = pd.read_csv(filepath)  # read the data
```

```python
# Let's explore the dataframe using .columns and .describe()
print(df.columns)

print()

# .describe()
# count, mean, std. min, 25%, 50% (median), 75%, max
print(df.describe())
```

```python
# Sort the dataframe by Rating, best one first!
df.sort_values('Rating', ascending=False)
# inplace does not modify the actual dataframe
```

```python
# Find out if Year and Rating are correlated
correlation = df['Year'].corr(df['Rating'])
print(f"Correlation between Year and Rating: {correlation}")
```

```python
# Plot the year and rating as a scatter plot
import matplotlib.pyplot as plt

# Scatter plot of Year vs. Rating
plt.scatter(df['Year'], df['Rating'], alpha=0.5)
plt.title('Year vs Rating Scatter Plot')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.show()
```

```python
# Task 1: Count & average rating of all movies
movie_count = df['Rating'].count()
average_rating = df['Rating'].mean()
print(movie_count, average_rating)
```

```python
# Task 2: Number & average rating of all movies with more than 1000 votes
movies_with_more_votes = df[df['Votes'] > 1000]
movies_with_more_votes_count = movies_with_more_votes['Rating'].count()
average_rating_more_votes = movies_with_more_votes['Rating'].mean()
print(movies_with_more_votes_count, average_rating_more_votes)
```

```python
# Task 3: Average rating of movies grouped by year
average_rating_by_year = df.groupby('Year')['Rating'].mean()
print(average_rating_by_year)
```

```python
# Task 5: Plot the average duration of movies per year as a line chart and scatter plot
average_duration_by_year = df.groupby('Year')['Duration (min)'].mean()

# Line chart for average duration per year
plt.figure(figsize=(10,6))
plt.plot(average_duration_by_year, marker='o')
plt.title("Average Duration of Movies Per Year (Line Chart)")
plt.xlabel("Year")
plt.ylabel("Average Duration (min)")
plt.grid(True)
plt.show()

# Scatter plot for average duration per year
plt.figure(figsize=(10,6))
plt.scatter(df['Year'], df['Duration (min)'], alpha=0.5)
plt.title("Average Duration of Movies Per Year (Scatter Plot)")
plt.xlabel("Year")
plt.ylabel("Duration (min)")
plt.show()
```

```python
# Task 6: Create a scatter plot of movie ratings vs number of votes
plt.figure(figsize=(10,6))
plt.scatter(df['Votes'], df['Rating'], alpha=0.5)
plt.title("Scatter Plot of Movie Ratings vs Number of Votes")
plt.xlabel("Number of Votes")
plt.ylabel("Rating")
plt.grid(True)
plt.show()
```

```python
# Task 7: Create a new column (Boolean) that indicates if a movie is a horror movie
df['IsHorror'] = df['Genre'].str.contains('Horror', case=False, na=False)
```

```python
# Task 8: Create a new column (Boolean) that indicates if a movie is a comedy movie
df['IsComedy'] = df['Genre'].str.contains('Comedy', case=False, na=False)
```

```python
# Task 9: Compare the average rating of horror movies to the average rating of comedy movies
horror_avg_rating = df[df['IsHorror']]['Rating'].mean()
comedy_avg_rating = df[df['IsComedy']]['Rating'].mean()

# Display the results for horror and comedy movies
print(f"Total number of movies: {movie_count}")
print(f"Average rating of all movies: {average_rating:.2f}")
print(f"Number of movies with more than 1000 votes: {movies_with_more_votes_count}")
print(f"Average rating of movies with more than 1000 votes: {average_rating_more_votes:.2f}")
print(f"Average rating of horror movies: {horror_avg_rating:.2f}")
print(f"Average rating of comedy movies: {comedy_avg_rating:.2f}")
```
