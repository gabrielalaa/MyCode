```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "imdb-movies-dataset.csv"
df = pd.read_csv(filepath)  # read the data
```

```python
# Set the style for Seaborn plots
sns.set(style="whitegrid")
```

```python
# Scatter plot: Rating vs. Votes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Rating', data=df, alpha=0.5)
plt.title("Scatter Plot: Movie Ratings vs Votes")
plt.show()
```

```python
# Create a new column (Boolean) that indicates if a movie is a comedy movie
df['IsComedy'] = df['Genre'].str.contains('Comedy', case=False, na=False)
```

```python
# Box plot: Distribution of ratings grouped by genre (using IsComedy for simplicity)
plt.figure(figsize=(10, 6))
sns.boxplot(x='IsComedy', y='Rating', data=df)
plt.title("Box Plot: Rating Distribution (Comedy vs Others)")
plt.show()
```

```python
# Violin plot: Duration distribution for Comedy vs Others
plt.figure(figsize=(10, 6))
sns.violinplot(x='IsComedy', y='Duration (min)', data=df)
plt.title("Violin Plot: Duration Distribution (Comedy vs Others)")
plt.show()
```

```python
# Bar chart: Average rating for Comedy and Other movies
avg_rating_comedy_horror = df.groupby('IsComedy')['Rating'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='IsComedy', y='Rating', data=avg_rating_comedy_horror)
plt.title("Bar Chart: Average Rating for Comedy Movies vs Others")
plt.show()
```

```python
# Two plots side by side (scatter plot and box plot)
plt.figure(figsize=(15, 6))
```

```python
# Scatter Plot on the left
plt.subplot(1, 2, 1)
sns.scatterplot(x='Votes', y='Rating', data=df, alpha=0.5)
plt.title("Scatter Plot: Movie Ratings vs Votes")
```

```python
# Box Plot on the right
plt.subplot(1, 2, 2)
sns.boxplot(x='IsComedy', y='Rating', data=df)
plt.title("Box Plot: Rating Distribution (Comedy vs Others)")

plt.tight_layout()
plt.show()
```
